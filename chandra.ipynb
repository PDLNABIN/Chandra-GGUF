```python
## Download Models

Download both models from Hugging Face:

1. [chandra-ocr-2.Q8_0.gguf](https://huggingface.co/prithivMLmods/chandra-ocr-2-GGUF/blob/main/chandra-ocr-2.Q8_0.gguf)
2. [chandra-ocr-2.mmproj-f32.gguf](https://huggingface.co/prithivMLmods/chandra-ocr-2-GGUF/blob/main/chandra-ocr-2.mmproj-f32.gguf)

You can use `huggingface-hub` or download them manually:
```

```python
from huggingface_hub import hf_hub_download
from pathlib import Path

TARGET_DIR = Path('/root/chandra_gguf')
TARGET_DIR.mkdir(parents=True, exist_ok=True)
FILES = [
    ('prithivMLmods/chandra-ocr-2-GGUF', 'chandra-ocr-2.Q8_0.gguf'),
    ('prithivMLmods/chandra-ocr-2-GGUF', 'chandra-ocr-2.mmproj-f32.gguf'),
]

for repo_id, filename in FILES:
    path = hf_hub_download(repo_id=repo_id, filename=filename, local_dir=str(TARGET_DIR), local_dir_use_symlinks=False)
    print(f'Downloaded {filename} -> {path}')
```

```python
# Install dependencies
apt-get update && apt-get install -y cmake build-essential git

# Clone llama.cpp
git clone https://github.com/ggml-org/llama.cpp
cd llama.cpp

# Build with CUDA + LLAVA support
cmake -B build \
  -DGGML_CUDA=ON \
  -DLLAMA_CURL=OFF

cmake --build build --config Release -j$(nproc)
```

```python
from huggingface_hub import hf_hub_download
from pathlib import Path

TARGET_DIR = Path('/root/chandra_gguf/models')
TARGET_DIR.mkdir(parents=True, exist_ok=True)
FILES = [
    ('prithivMLmods/chandra-ocr-2-GGUF', 'chandra-ocr-2-Q4_K_M.gguf'),
    ('prithivMLmods/chandra-ocr-2-GGUF', 'chandra-ocr-2.mmproj-q8_0.gguf'),
]

for repo_id, filename in FILES:
    path = hf_hub_download(repo_id=repo_id, filename=filename, local_dir=str(TARGET_DIR), local_dir_use_symlinks=False)
    print(f'Downloaded {filename} -> {path}')
```

```python
./build/bin/llama-server \
  --model /root/chandra_gguf/models/chandra-ocr-2-Q4_K_M.gguf \
  --mmproj /root/chandra_gguf/models/chandra-ocr-2.mmproj-q8_0.gguf \
  --n-gpu-layers 32 \
  --ctx-size 30000 \
  --host 0.0.0.0 \
  --port 8000
```
